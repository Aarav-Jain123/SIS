```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
import os
import requests
import time
```

```python
df = pd.read_excel('weatherdata.xlsx',)
df
```

```python
Numerics = LabelEncoder()
inputs = df.drop('Play', axis='columns')
target = df['Play']
target
```

```python
inputs['outloook_n']=Numerics.fit_transform(inputs['Outlook'])
inputs['temp_n']=Numerics.fit_transform(inputs['Temp'])
inputs['humidity_n']=Numerics.fit_transform(inputs['Humidity'])
inputs['windy_n']=Numerics.fit_transform(inputs['Windy'])
inputs
```

```python
inputs_n = inputs.drop(['Outlook', 'Temp', 'Humidity', 'Windy'], axis='columns')
inputs_n
```

```python
cr = GaussianNB()
cr.fit(inputs_n, target)
```

```python
cr.score(inputs_n, target)
```

```python
def predict_value_and_post():
    s = list(cr.predict([[1,1,0,1]]))
    res = {"data": s}
    r = requests.post('http://127.0.0.1:8000/api/weather/', data=res)
    
    return [res, r.status_code]

# while True: 
predict_value_and_post()
time.sleep(42)
```
